```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,10, 51)
```

```python
f = [np.cos(el) for el in t]
```

```python
plt.plot(t,f, color = 'green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.xlim([0.5,9.5])
plt.ylim([-2.5, 2.5])
```
